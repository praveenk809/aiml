```python
# Linear, Cubic, Log, Exp (Subplots)
# Linear
x = np.linspace(1,10,100)
y = 4*x
plt.figure(1) # row=1
plt.subplot(2,2,1) # grid 2x2, column=1
plt.title("Linear")
plt.xlabel("X")
plt.ylabel("Y")
plt.plot(x,y)

# Cubic
x = np.linspace(1,10,100)
y = x**3
plt.figure(1) # row=1
plt.subplot(2,2,2) # grid 2x2, column=2
plt.title("Cubic")
plt.xlabel("X")
plt.ylabel("Y")
plt.plot(x,y)
plt.subplots_adjust(wspace=0.4)

# Exp
x = np.linspace(1,10,100)
y = np.exp(x)
plt.figure(2) # row=2
plt.subplot(2,2,1) # grid 2x2, column=1
plt.title("Exp")
plt.xlabel("X")
plt.ylabel("Y")
plt.plot(x,y)

# Log
x = np.linspace(1,10,100)
y = np.log(x)
plt.figure(2) # row=2
plt.subplot(2,2,2) # grid 2x2, column=1
plt.title("Log")
plt.xlabel("X")
plt.ylabel("Y")
plt.plot(x,y)
plt.subplots_adjust(wspace=0.4)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# read data
market_df = pd.read_csv("./global_sales_data/market_fact.csv")
customer_df = pd.read_csv("./global_sales_data/cust_dimen.csv")
product_df = pd.read_csv("./global_sales_data/prod_dimen.csv")
shipping_df = pd.read_csv("./global_sales_data/shipping_dimen.csv")
orders_df = pd.read_csv("./global_sales_data/orders_dimen.csv")

df_1 = pd.merge(market_df, customer_df, how='inner', on='Cust_id')
df_2 = pd.merge(df_1, product_df, how='inner', on='Prod_id')
df_3 = pd.merge(df_2, shipping_df, how='inner', on='Ship_id')
master_df = pd.merge(df_3, orders_df, how='inner', on='Ord_id')

master_df.head()
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(x="Customer_Segment", y="Profit", data=master_df[master_df.Profit>0], hue="Product_Category")
plt.yscale('log')
```

```python
sns.countplot(y="Product_Sub_Category", data=master_df)
```

```python
master_df['Order_Date'] = pd.to_datetime(master_df['Order_Date'])
master_df.info()
```

```python
time_df = master_df.groupby('Order_Date')['Sales'].sum()
time_df
```

```python
plt.figure(figsize=(16,8))
sns.lineplot(data=time_df)
```
