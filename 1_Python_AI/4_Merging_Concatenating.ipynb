```python
import pandas as pd
market_df = pd.read_csv("global_sales_data/market_fact.csv")
market_df
```

```python
customer_df = pd.read_csv("global_sales_data/cust_dimen.csv")
customer_df
```

```python
product_df = pd.read_csv("global_sales_data/prod_dimen.csv")
product_df
```

```python
shipping_df = pd.read_csv("global_sales_data/shipping_dimen.csv")
shipping_df
```

```python
orders_df = pd.read_csv("global_sales_data/orders_dimen.csv")
orders_df
```

```python
df_1 = pd.merge(market_df,customer_df, on="Cust_id", how="inner")
df_2 = pd.merge(df_1,product_df, on="Prod_id", how="inner")
df_3 = pd.merge(df_2,shipping_df, on="Ship_id", how="inner")
master_df = pd.merge(df_3,orders_df, on="Ord_id", how="inner")
master_df.head()
```
