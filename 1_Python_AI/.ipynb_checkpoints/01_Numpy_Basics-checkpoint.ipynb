```python
import numpy as np

array_1 = np.arange(12).reshape(3,4)
array_2 = np.arange(15).reshape(3,5)

print(array_1)
print('\n')
print(array_2)
```

```python
result = np.hstack((array_1,array_2))
result
```

```python
result.shape
```

```python
array_1 = np.arange(12).reshape(3,4)
array_2 = np.arange(20).reshape(5,4)
print(array_1)
print('\n')
print(array_2)
```

```python
result = np.vstack((array_1,array_2))
result
```

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print(a)
print(b)
```

```python
result = np.stack((a,b), axis=0)
result
```

```python
result = np.stack((a,b), axis=1)
result
```
