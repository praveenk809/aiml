```python
import pandas as pd
df1 = pd.DataFrame({
    "ID": [1,2,3,4],
    "Name": ['Alice', 'Bob', 'Charlie', 'David']
})
df1
```

```python
df2 = pd.DataFrame({
    'ID': [3, 4, 5, 6],
    'Age': [25, 30, 35, 40],
})
df2
```

```python
inner_join = pd.merge(df1,df2, on="ID", how="inner")
inner_join
```

```python
left_join = pd.merge(df1,df2, on="ID", how="left")
left_join
```

```python
right_join = pd.merge(df1,df2, on="ID", how="right")
right_join
```

```python
outer_join = pd.merge(df1,df2, on="ID", how="outer")
outer_join
```

```python
df1 = pd.DataFrame({'Name': ['Aman', 'Joy', 'Rashmi', 'Saif'],
                    'Age': ['34', '31', '22', '33'],
                    'Gender': ['M', 'M', 'F', 'M']}
                  )
df1
```

```python
df2 = pd.DataFrame({'Name': ['Akhil', 'Asha', 'Preeti'],
                    'Age': ['31', '22', '23'],
                    'Gender': ['M', 'F', 'F']}
                  )
df2
```

```python
df3 = pd.concat([df1,df2], axis=0) # concat on rows
df3
```
